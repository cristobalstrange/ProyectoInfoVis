```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
brands_file = 'vis2/data/brand.csv'
brands_df = pd.read_csv(brands_file)

top10_brands = brands_df.nlargest(10, 'Total')

plt.figure(figsize=(16, 8))

base_color = 'lightblue'
highlighted_colors = [base_color] * len(top10_brands)
highlighted_colors[0] = 'darkcyan' 
highlighted_colors[-1] = 'darkcyan' 
bars = plt.barh(top10_brands['Brand'], top10_brands['Total'], color=highlighted_colors)

plt.gca().invert_yaxis()

plt.text(top10_brands['Total'].iloc[0] + 100000000, 0, f"${top10_brands['Total'].iloc[0]:,}", 
         va='center', fontsize=10, color='darkblue')

plt.text(top10_brands['Total'].iloc[-1] + 100000000, 9, f"${top10_brands['Total'].iloc[-1]:,}", 
         va='center', fontsize=10, color='darkblue')

plt.xlabel('Total Recaudado en Taquilla (Billones)')
plt.title('Top 10 productoras cinematogr√°ficas a nivel mundial')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
