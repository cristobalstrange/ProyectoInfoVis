```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_con_las_mayores_recaudaciones"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
tabla_pelis_mayor_recaudacion_anho = soup.find_all("table")[2]
```

```python
df_pelis_mayor_recaudacion = pd.read_html(tabla_pelis_mayor_recaudacion_anho.prettify())[0]
df_pelis_mayor_recaudacion
```

```python
df_pelis_mayor_recaudacion.info()
```

```python
# Cambiamos el nombre de la columna de "Recaudación mundial (USD)" a "Recaudacion mundial (Millones) (USD)"
df_pelis_mayor_recaudacion.rename(columns={"Recaudación mundial (USD)": "Recaudacion mundial (Millones) (USD)"}, inplace=True)

# Cambiamos los valores a int
df_pelis_mayor_recaudacion["Recaudacion mundial (Millones) (USD)"] = df_pelis_mayor_recaudacion["Recaudacion mundial (Millones) (USD)"].str.replace("$", "").str.replace("\xa0", "").str.replace(" ", "").astype(int)
```

```python
df_pelis_mayor_recaudacion
```

```python
# Tomamos la peliculas con mayor recaudación de cada productora
peliculas_mayor_recaudacion_productora = df_pelis_mayor_recaudacion.loc[df_pelis_mayor_recaudacion.groupby('Productora')['Recaudacion mundial (Millones) (USD)'].idxmax()].sort_values(by="Año de estreno", ascending=False)
peliculas_mayor_recaudacion_productora
```

```python
df_pelis_mayor_recaudacion.to_csv("peliculas_mayor_recaudacion.csv", index=False)
```
