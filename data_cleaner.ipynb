# Top 5 Géneros

```python
d = dict()
for genres in df["genres"].dropna():
    try:
        genres = genres.split(" ")
    except:
        print(genres)
        continue
    for g in genres:
        if g not in d.keys():
            d[g] = 1
        else:
            d[g] += 1

a = []
for g in d.keys():
    a.append((g,d[g]))
    
a.sort(key = lambda x:x[1])
a[-6:-1]
```
