```python
d = dict()
for genres in df["genres"].dropna():
    try:
        genres = genres.split(" ")
    except:
        print(genres)
        continue
    for g in genres:
        if g not in d.keys():
            d[g] = 1
        else:
            d[g] += 1

a = []
for g in d.keys():
    a.append((g,d[g]))
    
a.sort(key = lambda x:x[1])
a[-6:-1]
```

```python
def encontrar_decada(x):
    # print(x)
    intervalo = 5
    año = int(x["year"])
    decadas = (range(1910,2020,intervalo))
    for d in decadas:
        if año >= d and año<d+intervalo:
            return d
            
```

```python
df = pd.read_csv("data/movie_dataset.csv")
df = df[["genres","original_title","popularity","release_date"]]
df = df.dropna()

df["year"] = df.apply(lambda x : str(x["release_date"]).split("-")[0], axis=1)
print("max:",df["year"].max(), "min:",df["year"].min())
df["year"] = df.apply(encontrar_decada, axis=1)
df.head()
```

```python
df[df["year"] == 1942].sort_values(by="popularity", ascending=False)
```

```python
df["action"] = df.apply(lambda x:"Action" in str(x["genres"]).split(" "), axis=1)
df["adventure"] = df.apply(lambda x:"Adventure" in str(x["genres"]).split(" "), axis=1)
df["romance"] = df.apply(lambda x:"Romance" in str(x["genres"]).split(" "), axis=1)
df["comedy"] = df.apply(lambda x:"Comedy" in str(x["genres"]).split(" "), axis=1)
df["thriller"] = df.apply(lambda x:"Thriller" in str(x["genres"]).split(" "), axis=1)
df.head()
```

```python
df[df["year"] == 1960].sort_values(by="popularity", ascending=False)
```

```python
df[df["year"] == 1960].sort_values(by="popularity", ascending=False)
```

```python
means_action = df[df.action].groupby("year")["popularity"].mean()
means_action = pd.DataFrame(means_action)
means_action["genre"] = ["action" for i in range(len(means_action))]

means_adv = df[df.adventure].groupby("year")["popularity"].mean()
means_adv = pd.DataFrame(means_adv)
means_adv["genre"] = ["adventure" for i in range(len(means_adv))]

means_rom = df[df.romance].groupby("year")["popularity"].mean()
means_rom = pd.DataFrame(means_rom)
means_rom["genre"] = ["romance" for i in range(len(means_rom))]

means_comedy = df[df.comedy].groupby("year")["popularity"].mean()
means_comedy = pd.DataFrame(means_comedy)
means_comedy["genre"] = ["comedy" for i in range(len(means_comedy))]

means_thriller = df[df.thriller].groupby("year")["popularity"].mean()
means_thriller = pd.DataFrame(means_thriller)
means_thriller["genre"] = ["thriller" for i in range(len(means_thriller))]
means_thriller.head()
```

```python
means = pd.concat([means_thriller,means_comedy,means_rom,means_adv,means_action])
means.head()
```

```python
import seaborn as sns
```

```python
sns.set_theme(rc={'figure.figsize':(16,9)})
sns.lineplot(means,y="popularity",x="year",hue="genre")
```

# "Outliers"

```python
df[df["year"]<=2000].sort_values(by="popularity", ascending=False)
```

```python
maximos = []
df = df[df["year"]<=1990].sort_values(by="popularity", ascending=False)
for i in ["action","adventure","romance","comedy","thriller"]:
    a = df[df[i]]
    maxi = a["popularity"].max()
    a = df[df["popularity"] == maxi]
    maximos.append([i,str(a.loc[:,"original_title"].iloc[0]),int(a.loc[:,"popularity"]),int(a.loc[:,"year"])])
maximos
```

```python
sns.set_theme(rc={'figure.figsize':(16,9)})
ax = sns.lineplot(means[means.index<= 2000],y="popularity",x="year",hue="genre")
n = 5
peliculas_repetidas = []
for maximo in maximos:
    if maximo[1] in peliculas_repetidas:
        continue
    ax.axvline(x = maximo[3],    # Line on x = 2
           ymin = 0, # Bottom of the plot
           ymax = 1,ls="--") # Top of the plot
    ax.text(x=maximo[3],y=0+n,s=maximo[1])
    peliculas_repetidas.append(maximo[1])
    n+= 5
```
